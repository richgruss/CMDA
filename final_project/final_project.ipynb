# Team 10 Final Project: Predicting Bike Sharing Demand
Kaggle Competition:  Given the time of day, the day of the week, and 9 other features of a certain day, predict how many total bicycles will be rented for that hour.  See  see https://www.kaggle.com/c/bike-sharing-demand

There is significant demand for this kind of data modeling, as there are over 500 bike sharing programs throughout the world.

The features: 
 - season (categ., 1=spring, etc.)
 - holiday (bool., 0=not holiday, 1=holiday)
 - workingday (bool., 0=not workday, 1=workday)
 - weather (categ., 1-4 based on precip)
 - temp (num., temp in celsius)
 - atemp (num., "feels like" temp)
 - humidity (num., relative humidity)
 - windspeed (num., kph)
 


## 1.  Importing Data and Saving as Pickle

```python
%matplotlib inline
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import random

#read bicycle data into pandas dataframe: 
bike_data_orig = pd.read_csv('kaggle_bike_sharing_train.csv', parse_dates=[0], infer_datetime_format=True)
print bike_data_orig.shape
```

## 2. Reshaping Data

We want to use time of day as a feature, independent of date, so split out time, use as an integer and drop datetime, reording the columns we want.

```python
# create a new dataframe that drops the datetime and uses time as an int
bike_data_with_time = bike_data_orig.copy()

temp = pd.DatetimeIndex(bike_data_orig['datetime'])
bike_data_with_time['date'] = temp.date
bike_data_with_time['time'] = temp.time

bike_data_with_time['time_int'] = bike_data_with_time.apply(lambda row: int(row['time'].hour), axis=1)

#reorder columns to have date and time at the top again
bike_data_with_time=bike_data_with_time[['time_int','season', 'holiday', 'workingday', 'weather',
                     'atemp', 'humidity', 'windspeed', 'count']]

bike_data_with_time.head()
```

## 3. Treatment for Missing Values.

We don't have any missing data within the existing rows, possibly because rows were already removed.
For example, Jan. 4 is missing from 1:00pm to 4:00pm rows entirely. We are also missing some days.
However, we would like to create daily aggregates, in case bikes are provisioned on a daily basis, and this presents a missing data problem: "NA" rows are created for any missing days during aggregation.
Aggregate by day, dropping any missing days.

```python
bike_data_daily = bike_data_orig.copy()

bike_data_daily.index=pd.to_datetime(bike_data_daily.datetime,unit='H')

bike_data_daily = bike_data_daily.resample('D',how={'count': [np.sum,np.mean], 'season': np.mean,
                                                'holiday': np.mean, 'workingday': np.mean,
                                                'weather': [np.mean, np.std], 'temp': [np.mean, np.std],
                                                'atemp': [np.mean, np.std],
                                                'humidity': [np.mean, np.std], 'windspeed': [np.mean, np.std], })



# This seems to resample based on what we need but puts in missing values
# -- NAN -- for the days we don't have data for

# We can fix that by dropping NA's

bike_data_daily.resample('D').dropna()

bike_data_daily.head()
```

Since we would like a numerical prediction for bike demand given a vector of features, we'll try a linear regression first.

```python
def linearRegression(bike_data):
    bike_data_X = bike_data.ix[:, 0:7]
    bike_data_y = bike_data.ix[:, 8]

    X_train, X_test, y_train, y_test = train_test_split(bike_data_X, bike_data_y, test_size=0.8)

    # Create linear regression object
    regr = linear_model.LinearRegression(normalize=True)

    # Train the model using the training sets
    regr.fit(X_train, y_train)

    # The coefficients
    print('Coefficients: \n', regr.coef_)
    # The mean square error
    print("Residual sum of squares: %.2f"
          % np.mean((regr.predict(X_test) - y_test) ** 2))
    # Explained variance score: 1 is perfect prediction
    print('Variance score: %.2f' % regr.score(X_test, y_test))

    test_dataframe = pd.DataFrame()
    test_dataframe['actual_count'] = y_test
    test_dataframe['pred_count'] = regr.predict(X_test)

    test_dataframe = test_dataframe.sort(columns=['pred_count'])

    #print test_dataframe

    plt.scatter(test_dataframe['pred_count'], test_dataframe['actual_count'],  color='black')

    plt.xlabel('Predicted Hourly Rentals')
    plt.ylabel('Actual Hourly Rentals')

    plt.show()
```

### Algorithm 2: Naive Bayes Classification
Our next attempt is a Naive Bayes classifier, which means we'll have to bin our target variables
into categories.  Fewer variables will give more accurate predictions, so we'll use 5 bins.  This will give users a precition range of within 200 bikes out of 1000. 

```python
#utility functions for binning and model validation
def categorize(bins, val):
    for idx,item in enumerate(bins):
        if idx == len(bins)-1: #we got to the end
            return idx
        elif val >= item and val < bins[idx+1]:
            return idx
    return -1 #not found.  shouldn't happen

def evaluate_cross_validation(clf,X,y,K):
   cv = KFold(len(y), K, shuffle=True, random_state=0)
   scores = cross_val_score(clf, X, y, cv=cv)
   print scores
   print ("Mean score: {0:.3f} (+/- {1:.3f})").format(np.mean(scores), sem(scores))
```

```python
def naiveBayesClassify(bike_data):
     #scale numeric features
    min_max_scaler = preprocessing.MinMaxScaler()
    bike_data['atemp_scaled'] = min_max_scaler.fit_transform(bike_data['atemp'])
    bike_data['humidity_scaled'] = min_max_scaler.fit_transform(bike_data['humidity'])
    bike_data['windspeed_scaled'] = min_max_scaler.fit_transform(bike_data['windspeed'])

    #bin according to count
    bins = np.linspace(bike_data['count'].min(), bike_data['count'].max(), 5)
    #bins is an ndarray where value is min for that bin
    bike_data['category'] = bike_data.apply(lambda row: categorize(bins,row['count']), axis=1)


    bike_data=bike_data[['time_int','season', 'holiday', 'workingday', 'weather',
                         'atemp_scaled', 'humidity_scaled', 'windspeed_scaled', 'category']]


    #print bike_data[bike_data['count'] < 10]

    print bike_data

    bike_data_X = bike_data.ix[:, 0:7]
    bike_data_y = bike_data.ix[:, 8]

    X_train, X_test, y_train, y_test = train_test_split(bike_data_X, bike_data_y, test_size=0.8)

    clf = MultinomialNB()
    clf.fit(X_train, y_train)
    MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)

    test_dataframe = pd.DataFrame()
    test_dataframe['category'] = y_test
    test_dataframe['pred_category'] = clf.predict(X_test)

    print test_dataframe


    evaluate_cross_validation(clf, X_train, y_train, 5)
    evaluate_cross_validation(clf, X_test, y_test, 5)
```
